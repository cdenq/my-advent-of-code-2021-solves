# Data Clean

```python
#importing data
with open('puzzle_input.txt', 'r') as input_data:
    input_list = [line.strip() for line in input_data]
```

```python
#create vars to hold info
output_ls = []
temp_x_max, temp_y_max = 0, 0

#loop through input
for i in input_list:
    #create / reset the temp_ls to hold information on each line
    temp_ls = []

    #adding xy coordinates of each line into temp_ls
    for j in i.split()[0].split(','):
        j = int(j)
        temp_ls.append(j)
        if j > temp_x_max:
            temp_x_max = j
    for k in i.split()[2].split(','):
        k = int(k)
        temp_ls.append(k)
        if k > temp_y_max:
            temp_y_max = k

    #adding type of line into temp_ls
    if (temp_ls[0] == temp_ls[2]) & (temp_ls[1] == temp_ls[3]):
        temp_ls.append('P') #if both xy are same, then its a point
    elif (temp_ls[0] == temp_ls[2]):
        temp_ls.append('V') #else if x same, then its vertical line
    elif (temp_ls[1] == temp_ls[3]):
        temp_ls.append('H') #else if y same, then its horizontal line
    else:
        temp_ls.append('D') #else its a diagnol line

    #adding finished temp_ls to output_ls
    output_ls.append(temp_ls)
```
